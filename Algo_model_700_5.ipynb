```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load dataset
data = pd.read_csv('C://Users//Jayashrinidhi V//OneDrive//Documents//VScode//AlgoMaster//data_aes_des_3des.csv')
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Preprocess the data
# Encode categorical labels (algorithm)
label_encoder = LabelEncoder()
data['algorithm'] = label_encoder.fit_transform(data['algorithm'])
```

```python

# Convert text columns to numerical features
vectorizer = TfidfVectorizer()
```

```python

# Apply TF-IDF to text columns
plaintext_features = vectorizer.fit_transform(data['plaintext'])
key_features = vectorizer.fit_transform(data['key'])
iv_features = vectorizer.fit_transform(data['iv'])
encrypted_message_features = vectorizer.fit_transform(data['encrypted_message'])
```

```python

# Combine all features into a single feature matrix
from scipy.sparse import hstack
X = hstack([plaintext_features, key_features, iv_features, encrypted_message_features])

# Target variable
y = data['algorithm']
```

```python


# Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize classifiers
svm_classifier = SVC(kernel='linear')
knn_classifier = KNeighborsClassifier(n_neighbors=5)
naive_bayes_classifier = GaussianNB()
```

```python

# Train classifiers
svm_classifier.fit(X_train, y_train)
knn_classifier.fit(X_train, y_train)
naive_bayes_classifier.fit(X_train.toarray(), y_train)  
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
"""
# Define the parameter grid
#Accuracy: 0.1838095238095238

param_grid_svm = {
    'C': [0.01, 0.1, 1],
    'kernel': ['linear', 'rbf'],  # Radial Basis Function kernel
    'gamma': [0.0001, 0.001, 0.01, 0.1]  # Only used if kernel='rbf'
}"""

param_grid_svm = {
    'C': [0.01, 0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': [0.0001, 0.001, 0.01, 0.1, 1]  # Applicable for 'rbf' and 'poly'
    }

# Initialize GridSearchCV for SVM
grid_search_svm = GridSearchCV(SVC(), param_grid_svm, cv=5, n_jobs=-1, verbose=1)

# Train GridSearchCV
grid_search_svm.fit(X_train, y_train)

# Best parameters and model
print("Best parameters for SVM:", grid_search_svm.best_params_)
best_svm_model = grid_search_svm.best_estimator_

# Predict and evaluate
svm_predictions = best_svm_model.predict(X_test)
print("SVM Classifier:")
print("Accuracy:", accuracy_score(y_test, svm_predictions))
print("Classification Report:\n", classification_report(y_test, svm_predictions))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]  # Power parameter for the Minkowski distance
}

# Initialize GridSearchCV for KNN
grid_search_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=5, n_jobs=-1, verbose=1)

# Train GridSearchCV
grid_search_knn.fit(X_train, y_train)

# Best parameters and model
print("Best parameters for KNN:", grid_search_knn.best_params_)
best_knn_model = grid_search_knn.best_estimator_

# Predict and evaluate
knn_predictions = best_knn_model.predict(X_test)
print("KNN Classifier:")
print("Accuracy:", accuracy_score(y_test, knn_predictions))
print("Classification Report:\n", classification_report(y_test, knn_predictions))
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for MultinomialNB
param_grid_nb = {
    'alpha': [0.1, 0.5, 1.0, 1.5]  # Only applicable for MultinomialNB
}

# Initialize GridSearchCV for MultinomialNB
grid_search_nb = GridSearchCV(MultinomialNB(), param_grid_nb, cv=5, n_jobs=-1, verbose=1)

# Train GridSearchCV
grid_search_nb.fit(X_train.toarray(), y_train)  # Use toarray() for Naive Bayes

# Best parameters and model
print("Best parameters for Naive Bayes:", grid_search_nb.best_params_)
best_nb_model = grid_search_nb.best_estimator_

# Predict and evaluate
naive_bayes_predictions = best_nb_model.predict(X_test.toarray())
print("Naive Bayes Classifier:")
print("Accuracy:", accuracy_score(y_test, naive_bayes_predictions))
print("Classification Report:\n", classification_report(y_test, naive_bayes_predictions))
```
